```python
df=pd.read_csv('Churn_Modelling (1).csv')
```

```python
df.head()
```

```python
df=df.drop(['RowNumber','CustomerId','Surname'],axis=1)
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
label_enccoder=LabelEncoder()
df['Gender']=label_enccoder.fit_transform(df['Gender'])
```

```python
df.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
one_encoded=OneHotEncoder()
geo_encoded=one_encoded.fit_transform(df[['Geography']])
```

```python
geo_encoded
```

```python
one_encoded.get_feature_names_out(['Geography'])
```

```python
geo_encoded_df=pd.DataFrame(geo_encoded.toarray(),columns=one_encoded.get_feature_names_out(['Geography']))
```

```python
geo_encoded_df
```

```python
df=pd.concat([df.drop(['Geography'],axis=1),geo_encoded_df],axis=1)
```

```python
df.head()
```

```python
import pickle
```

```python
with open('label_encoded.pkl','wb') as file:
    pickle.dump(label_enccoder,file)
with open('Hot_encoded.pkl','wb') as file:
    pickle.dump(one_encoded,file)
```

```python
x=df.drop(['Exited'],axis=1)
y=df['Exited']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
with open('scaler.pkl','wb') as file:
    pickle.dump(scaler,file)
```

```python
%pip install tensorflow
## create a model
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard
import datetime
```

```python
x_train.shape[1]
```

```python
model=Sequential([
    Dense(64,activation='relu',input_shape=(x_train.shape[1],)),
    Dense(32,activation='relu'),
    Dense(1,activation='sigmoid')
])
```

```python
model.summary()
```

```python
##compile 
model.compile(optimizer='Adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.save("model.h5")
```

```python
log_dir = "logs/fit/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
```

```python
tensorboard_callback = TensorBoard(
    log_dir=log_dir,
    histogram_freq=1
)
```

```python
tensorboard_callback
```

```python
early_stopping_callback=EarlyStopping(
    monitor='val_loss',
    patience=5,
    restore_best_weights=True
)
```

```python
early_stopping_callback
```

```python
##train the model with tensorboard
training=model.fit(
    x_train,
    y_train,
    epochs=100,
    validation_data=(x_test,y_test),
    callbacks=[tensorboard_callback,early_stopping_callback]
)
```
